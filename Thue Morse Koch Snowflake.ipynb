```python
from tkinter import *
import turtle
import numpy as np
```

```python
#generate the boolean complement of a string

def complement(string):
    comp_string = ""
    
    for i in range(len(string)):
        if (string[i] == '0'):
            comp_string += '1'
            
        else:
            comp_string += '0'
            
    return comp_string

#generate the nth thue-morse number
def t(n,s):
    
    
    
    for i in range(n-1):
        s += complement(s)
        
    return s

def main():
    turtle.speed(0)
    test = t(15,"0")
    print(test)
    
    for m in range(len(test)):
        if (test[m] == '0'):
            turtle.forward(2)
        else:
            turtle.right(60)
            
main()

        
```
